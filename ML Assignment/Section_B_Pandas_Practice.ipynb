```python
# 1. Create a DataFrame of 5 students with columns: Name, age, Marks. Find the students with the highest marks
import pandas as pd
import numpy as np
df1 = pd.DataFrame({
    'Name': ['sid', 'aman', 'ansaf', 'jeetan', 'deba'],
    'Age': [24, 25, 23, 25, 24],
    'Marks': [88, 97, 90, 36, 97]
})
df1
```

```python
# find the students with the highest marks
highest_marks = df1['Marks'].max()
student_names = df1.loc[df1['Marks'] == highest_marks]
print("Complete details of students with the highest marks:\n", student_names)

# only student names (no other columns)
print("The students with the highest marks:\n", student_names['Name'])
```

```python
# 3. Add a new column PASS/FAIL with marks>= 40, otherwise

df1['Result'] = np.where(df1['Marks'] >= 40, 'PASS', 'FAIL')

# using list comprehension
df1['Result2'] = ['PASS' if m >= 40 else 'FAIL' for m in df1['Marks']]

print(df1)
```

```python
# 4. Group the data by a categorial column and compute the average of numeric columns: marks and age
df1_group = df1.groupby('Result')
df1_group.mean(['marks', 'age'])
```

```python
# 5. Merge two DataFrames: one containing Employee IDs and Names, another containing Employee IDs and Salaries
df_emp1 = pd.DataFrame({
    'Employee_ID': [1, 2, 3, 4, 5, 6],
    'Emp_Names': ['sid', 'aman', 'ansaf', 'jeetan', 'deba', 'sid']
})
df_emp2 = pd.DataFrame({
    'Employee_ID': [1, 2, 3, 4, 5, 6],
    'Salaries (in K)': [35, 15, 10, 40, 37, 30]
})
merge_df = pd.merge(df_emp1, df_emp2, on='Employee_ID')
print(merge_df)
```

```python
# 6. Sort the DataFrame by multiple columns: first by Department, then by Salary (descending)
merge_df['Department'] = ['IT', 'HR', 'Manager', 'HR', 'IT', 'IT']
merge_df = merge_df.sort_values(by=['Emp_Names', 'Salaries (in K)'], ascending=[True, False])
print(merge_df)
```

```python
# 7. Replace all missing values in a DataFrame with column mean
# create a df7 with missing values
df7 = pd.DataFrame({
    'A': [1, np.nan, 4, np.nan, 5],
    'B': [60, 70, 81, 79, np.nan],
})
df7.fillna(np.round(df7.mean(),2), inplace=True)
print(df7)

# check for na (missing value) if yes then fill with mean using list comprehension
df7_2 = pd.DataFrame({
    'A': [1, np.nan, 4, np.nan, 5],
    'B': [60, 70, 81, 79, np.nan],
})

df7_2['A'] = [np.round(df7_2['A'].mean(),2) if np.isnan(x) else x for x in df7_2['A']]
df7_2['B'] = [np.round(df7_2['B'].mean(),2) if np.isnan(x) else x for x in df7_2['B']]
print(df7_2)
```

```python
# 8. Create a new coloumn that shows the rank of employee by salary within their department
merge_df['Rank'] = merge_df.groupby('Department')['Salaries (in K)'].rank('dense', ascending=False)
print(merge_df)

merge_df = merge_df.sort_values(by=['Department', 'Rank'])
print('\n', merge_df)
```

```python
# 9. Use apply() to increase every salary by 15%
merge_df['Salaries (in K)'] = merge_df['Salaries (in K)'].apply(lambda x: x * 1.15)
print(merge_df)
```

```python
# 10. Convert a date column into day, month, and year components
# add date column in merge_df
merge_df['Date'] = pd.to_datetime(['2023-01-01', '2020-02-03', '2023-05-11', '2023-07-26', '2024-05-21', '2021-11-11'])

# adding day, month, year seperate column from Date column
merge_df['Day'] = merge_df['Date'].dt.day
merge_df['Month'] = merge_df['Date'].dt.month
merge_df['Year'] = merge_df['Date'].dt.year
print(merge_df)
```
