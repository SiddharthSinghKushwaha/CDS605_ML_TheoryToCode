## Matplotlib Practice problems

```python
# importing libraries
import matplotlib.pyplot as plt
import numpy as np

# 1. Plot sin(x) and cos(x) on the same graph for x belongs [0, 2PI] with legends, gridlines, and different colours
PI = np.pi
x = np.linspace(0, 2*PI, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('sin(x) and cos(x)')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Create a bar chart of 5 students (Math, English, Science, History, Art) with random student scores, Label axes properly
students = ['Math', 'English', 'Science', 'History', 'Art']
scores = np.random.randint(0, 100, 5)
plt.bar(students, scores)
plt.title('Student Scores')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.show()
```

```python
# 3. Generate a scatter plot of 50 random (x,y) points colored by their values using a colormap
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
plt.scatter(x, y, c=colors, cmap='viridis') # cmap for colourmap
plt.title('Scatter Plot of Random Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 4. Plot a histogram of 1000 random normal values with 40 bins. Change the color and add a grid.
values = np.random.normal(size=1000)  # normal will generates a normal distribution
plt.hist(values, bins=40, color='green')
plt.title('Histogram of Random Normal Values')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 5. Create 4 subplots showing (i) sin(x) (ii) cos(x) (iii) tanx(x) (iv) sin(x)+cos(x)
x = np.linspace(0, 2*PI, 100)

# Create a new x range for tan(x) that avoids the asymptotes (range of tan Generated using Gemini)
x_tan = np.linspace(0 + 0.1, 2*PI - 0.1, 100)
# Remove values close to the asymptotes
x_tan = x_tan[(x_tan % (PI/2) > 0.1) & (x_tan % (PI/2) < PI/2 - 0.1)]

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x_tan)
y4 = y1 + y2

fig, axs = plt.subplots(2, 2)

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('sin(x)')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('cos(x)')

plt.subplot(2, 2, 3)
plt.plot(x_tan, y3)
plt.title('tan(x)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('sin(x)+cos(x)')

plt.tight_layout()
plt.show()
```

```python
# 6. Plot a pie chart showing the percentage of time you spend on 4 daily activities (Sllep, Study, Exercise, Leisure)
activities = ['Sleep', 'Study', 'Exercise', 'Leisure', 'Other']
time_spent = [7, 11, 1, 2, 3]   # written in hour (will be convert into percentage)
time_spent = [x / sum(time_spent) * 100 for x in time_spent] # in percentage

plt.pie(time_spent, labels=activities, startangle=90, autopct='%1.1f%%')
plt.title('Time Spent on Daily Activities')
plt.show()
```

```python
# 7. Create a box plot comparing 3 datasets of random values with different variances
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(0, 2, 100)
data3 = np.random.normal(0, 3, 100)
plt.boxplot([data1, data2, data3], )
plt.title('Box Plot of Random Values')
plt.xticks([1, 2, 3], ['Data 1', 'Data 2', 'Data 3'])
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.show()
```

```python
# 8. Plot a 3D surface for z = sin(sqrt(x^2, y^2)) using Axes3D
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
