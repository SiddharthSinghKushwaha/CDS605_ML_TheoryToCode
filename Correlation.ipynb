```python
# Program to find corelation among features of the dataset (Class date: 14th Oct)
# import numpy pakage
import numpy as np
```

```python
# declare four features with 15 data values
f1 = np.array([10,11,10,12,11,10,13,10,12,11,10,11,10,12,11])
f2 = np.array([20,22,19,24,21,20,25,18,23,21,20,22,19,24,21])
f3 = np.array([5,6,5,7,6,5,8,5,7,6,5,6,5,7,6])
f4 = np.array([100,101,100,102,101,100,103,99,102,101,100,101,99,102,101])
```

```python
# calculating mean of each features
f1_mean = np.round(np.mean(f1), 2)
f2_mean = np.round(np.mean(f2), 2)
f3_mean = np.round(np.mean(f3), 2)
f4_mean = np.round(np.mean(f4), 2)
print(f1_mean, f2_mean, f3_mean, f4_mean)
```

```python
# subtracting each feature with its mean
dif_f1_f1mean = f1 - f1_mean
dif_f2_f2mean = f2 - f2_mean
dif_f3_f3mean = f3 - f3_mean
dif_f4_f4mean = f4 - f4_mean
print("Subtract f1 - f1mean", dif_f1_f1mean)
print("Subtract f2 - f2mean", dif_f2_f2mean)
print("Subtract f3 - f3mean", dif_f3_f3mean)
print("Subtract f4 - f4mean", dif_f4_f4mean)
```

```python
# squaring the prev. calculated differnce
sq_dif_f1_f1mean = dif_f1_f1mean ** 2
sq_dif_f2_f2mean = dif_f2_f2mean ** 2
sq_dif_f3_f3mean = dif_f3_f3mean ** 2
sq_dif_f4_f4mean = dif_f4_f4mean ** 2

# now calculting sum of these square
sum_sq_dif_f1_f1mean = np.round(np.sum(sq_dif_f1_f1mean), 2)
sum_sq_dif_f2_f2mean = np.round(np.sum(sq_dif_f2_f2mean), 2)
sum_sq_dif_f3_f3mean = np.round(np.sum(sq_dif_f3_f3mean), 2)
sum_sq_dif_f4_f4mean = np.round(np.sum(sq_dif_f4_f4mean), 2)

print(sum_sq_dif_f1_f1mean, sum_sq_dif_f2_f2mean, sum_sq_dif_f3_f3mean, sum_sq_dif_f4_f4mean)
```

```python
# product of dif_f1_f1_mean and dif_f2_f2_mean and other 5 more cases
product_f1_f2 = dif_f1_f1mean * dif_f2_f2mean
product_f1_f3 = dif_f1_f1mean * dif_f3_f3mean
product_f1_f4 = dif_f1_f1mean * dif_f4_f4mean
product_f2_f3 = dif_f2_f2mean * dif_f3_f3mean
product_f2_f4 = dif_f2_f2mean * dif_f4_f4mean
product_f3_f4 = dif_f3_f3mean * dif_f4_f4mean

# sum of these product of feature
sum_product_f1_f2 = np.round(np.sum(product_f1_f2), 2)
sum_product_f1_f3 = np.round(np.sum(product_f1_f3), 2)
sum_product_f1_f4 = np.round(np.sum(product_f1_f4), 2)
sum_product_f2_f3 = np.round(np.sum(product_f2_f3), 2)
sum_product_f2_f4 = np.round(np.sum(product_f2_f4), 2)
sum_product_f3_f4 = np.round(np.sum(product_f3_f4), 2)

print(sum_product_f1_f2, sum_product_f1_f3, sum_product_f1_f4, sum_product_f2_f3, sum_product_f2_f4, sum_product_f3_f4)
```

```python
# calculating correlation for 6 cases
corr_f1_f2 = np.round(sum_product_f1_f2 / np.sqrt(sum_sq_dif_f1_f1mean * sum_sq_dif_f2_f2mean), 2)
corr_f1_f3 = np.round(sum_product_f1_f3 / np.sqrt(sum_sq_dif_f1_f1mean * sum_sq_dif_f3_f3mean), 2)
corr_f1_f4 = np.round(sum_product_f1_f4 / np.sqrt(sum_sq_dif_f1_f1mean * sum_sq_dif_f4_f4mean), 2)
corr_f2_f3 = np.round(sum_product_f2_f3 / np.sqrt(sum_sq_dif_f2_f2mean * sum_sq_dif_f3_f3mean), 2)
corr_f2_f4 = np.round(sum_product_f2_f4 / np.sqrt(sum_sq_dif_f2_f2mean * sum_sq_dif_f4_f4mean), 2)
corr_f3_f4 = np.round(sum_product_f3_f4 / np.sqrt(sum_sq_dif_f3_f3mean * sum_sq_dif_f4_f4mean), 2)

print("Cor(F1, F2)", corr_f1_f2)
print("Cor(F1, F3)", corr_f1_f3)
print("Cor(F1, F4)", corr_f1_f4)
print("Cor(F2, F3)", corr_f2_f3)
print("Cor(F2, F4)", corr_f2_f4)
print("Cor(F3, F4)", corr_f3_f4)
```

## Above correlation code using GEMINI

Following is been generated by Gemini. I trying how it can perform the same task.

```python
import pandas as pd

data = {'f1': f1, 'f2': f2, 'f3': f3, 'f4': f4}
df = pd.DataFrame(data)
display(df.head())
```

```python
feature_means = df.mean()
print(feature_means)
```

```python
df_diff = df - feature_means
display(df_diff.head())
```

```python
df_diff_squared = df_diff ** 2
sum_of_squared_differences = df_diff_squared.sum()
print(sum_of_squared_differences)
```

```python
product_of_differences = df_diff.T.dot(df_diff)
sum_of_product_of_differences = product_of_differences.values
print(sum_of_product_of_differences)
```

```python
sqrt_sum_of_squared_differences = np.sqrt(sum_of_squared_differences)
outer_product_of_sqrt = np.outer(sqrt_sum_of_squared_differences, sqrt_sum_of_squared_differences)
correlation_matrix = sum_of_product_of_differences / outer_product_of_sqrt
print(correlation_matrix)
```

```python
correlation_matrix_df = pd.DataFrame(correlation_matrix, columns=df.columns, index=df.columns)
display(correlation_matrix_df)
```

Data Analysis Key Findings

*   The correlation matrix shows the pairwise correlation coefficients between each feature.
